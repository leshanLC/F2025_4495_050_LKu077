```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=UserWarning)
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# job skill datasets
job_skills_df = pd.read_csv('/content/sample_data/job_skills.csv')
job_meta = pd.read_csv('/content/sample_data/linkedin_job_postings.csv')
```

```python
# 2. job skills datasets merge
df_skills_merged = pd.merge(job_meta, job_skills_df, on="job_link", how="inner")


print(df_skills_merged.shape)
print(df_skills_merged.columns)
df_skills_merged.head(5)


# df_merged.to_csv('/content/sample_data/job_master.csv', index=False)
```

```python
# remove columns
drop_cols = ['job_location','last_processed_time', 'got_summary', 'got_ner',
             'is_being_worked', 'first_seen',
             'search_city', 'search_country', 'search_position']

df_skill_cleaned = df_skills_merged.drop(columns=drop_cols)

print(df_skill_cleaned.shape)
print(df_skill_cleaned.columns)
df_skill_cleaned.head(5)

# # save CSV file: job and skills 1st cleaned csv file
# df_skill_cleaned.to_csv('/content/sample_data/job_master_clean.csv', index=False)
```

I used two datasets from 1.3M linkedin jobs and skills -> merging two datasets and remove useless columns

handling null values

```python
#check the null value
df_skill_cleaned.isnull().sum()
```

```python
# remove null values from job_skill
# 1) blank/string -> 'nan'
def _norm_str(s):
    if pd.isna(s):
        return np.nan
    s = str(s).strip()
    return np.nan if s=="" or s.lower() in {"nan","none","null"} else s

for col in ["company","job_level","job_type","job_skills","job_title","job_link"]:
    df_skill_cleaned[col] = df_skill_cleaned[col].map(_norm_str)
```

```python
# 2) company null values -> unknown
df_skill_cleaned["company"] = df_skill_cleaned["company"].fillna("Unknown")
```

```python
# 3) remove blank columns in job_skills
#    - , seperate, remove 0 token as well
def split_skills(s):
    if pd.isna(s):
        return []
    toks = [t.strip().lower() for t in s.split(",") if t.strip()!=""]
    return toks

tmp_skills = df_skill_cleaned["job_skills"].apply(split_skills)
mask_valid_skills = tmp_skills.apply(lambda L: len(L) > 0)
df_skill_cleaned = df_skill_cleaned[mask_valid_skills].copy()
```

```python
# 4) skills column store as list (for modeling and graph)
df_skill_cleaned["skills_list"] = tmp_skills[mask_valid_skills]
```

```python
# 5) remove duplication by job_link keep latest/ first column
before = len(df_skill_cleaned)
df_skill_cleaned = df_skill_cleaned.drop_duplicates(subset=["job_link"], keep="first")
after = len(df_skill_cleaned)

print(f"remove duplicates before/ after: {before} -> {after}")
print(df_skill_cleaned.isna().sum())
df_skill_cleaned.head(5)
```

```python
# job_link duplicated columns check
dupes = df_skill_cleaned[df_skill_cleaned.duplicated(subset=["job_link"], keep=False)]
print(dupes.shape)
dupes.head(5)
```

```python
# set skills with lower case letters
df_skill_cleaned['job_skills'] = (
    df_skill_cleaned['job_skills']
    .str.lower()
    .str.replace('[^a-z, ]', '', regex=True)
)
```
