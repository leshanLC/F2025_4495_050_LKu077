```python
# remove null values from job_skill
# 1) blank/string -> 'nan'
def _norm_str(s):
    if pd.isna(s):
        return np.nan
    s = str(s).strip()
    return np.nan if s=="" or s.lower() in {"nan","none","null"} else s

for col in ["company","job_level","job_type","job_skills","job_title","job_link"]:
    df_skill_cleaned[col] = df_skill_cleaned[col].map(_norm_str)
```

```python
# 2) company null values -> unknown
df_skill_cleaned["company"] = df_skill_cleaned["company"].fillna("Unknown")
```

```python
# 3) remove blank columns in job_skills
#    - , seperate, remove 0 token as well
def split_skills(s):
    if pd.isna(s):
        return []
    toks = [t.strip().lower() for t in s.split(",") if t.strip()!=""]
    return toks

tmp_skills = df_skill_cleaned["job_skills"].apply(split_skills)
mask_valid_skills = tmp_skills.apply(lambda L: len(L) > 0)
df_skill_cleaned = df_skill_cleaned[mask_valid_skills].copy()
```

```python
# 4) skills column store as list (for modeling and graph)
df_skill_cleaned["skills_list"] = tmp_skills[mask_valid_skills]
```

```python
# 5) remove duplication by job_link keep latest/ first column
before = len(df_skill_cleaned)
df_skill_cleaned = df_skill_cleaned.drop_duplicates(subset=["job_link"], keep="first")
after = len(df_skill_cleaned)

print(f"remove duplicates before/ after: {before} -> {after}")
print(df_skill_cleaned.isna().sum())
df_skill_cleaned.head(5)
```

```python
# job_link duplicated columns check
dupes = df_skill_cleaned[df_skill_cleaned.duplicated(subset=["job_link"], keep=False)]
print(dupes.shape)
dupes.head(5)
```

```python
# set skills with lower case letters
df_skill_cleaned['job_skills'] = (
    df_skill_cleaned['job_skills']
    .str.lower()
    .str.replace('[^a-z, ]', '', regex=True)
)
```

```python
def categorize_job(title):
    """categorized by job title"""
    if pd.isna(title):
        return 'Other'

    title_lower = str(title).lower()

    categories = {
        'Data & Analytics': ['data', 'analyst', 'analytics', 'scientist', 'bi', 'business intelligence', 'database'],
        'Software Engineering': ['software', 'developer', 'engineer', 'programmer', 'full stack', 'frontend', 'backend', 'web developer'],
        'AI & ML': ['ai', 'machine learning', 'deep learning', 'nlp', 'computer vision', 'artificial intelligence'],
        'Cloud & DevOps': ['cloud', 'devops', 'aws', 'azure', 'gcp', 'infrastructure', 'system admin'],
        'Product Management': ['product manager', 'product owner', 'product management'],
        'Project Management': ['project manager', 'project management', 'scrum master', 'program manager'],
        'UX/UI Design': ['ux', 'ui', 'designer', 'user experience', 'user interface', 'graphic design'],
        'Marketing': ['marketing', 'digital marketing', 'growth', 'seo', 'sem', 'social media', 'content marketing'],
        'Sales': ['sales', 'account executive', 'business development', 'account manager', 'sales executive'],
        'Cybersecurity': ['security', 'cybersecurity', 'infosec', 'penetration testing', 'network security'],
        'Consulting': ['consultant', 'consulting', 'advisor', 'strategist'],
        'Finance': ['finance', 'financial', 'accounting', 'accountant', 'cfo', 'financial analyst'],
        'HR': ['hr', 'human resources', 'recruiter', 'talent acquisition'],
        'Healthcare': ['nurse', 'doctor', 'healthcare', 'medical', 'pharmacy', 'therapist'],
        'Education': ['teacher', 'professor', 'educator', 'instructor', 'trainer'],
        'Operations': ['operations', 'operational', 'logistics', 'supply chain'],
        'Customer Service': ['customer service', 'customer support', 'client service', 'help desk']
    }

    for category, keywords in categories.items():
        if any(keyword in title_lower for keyword in keywords):
            return category

    return 'Other'
```

```python
def integrated_strategic_sampling(df_skill_cleaned, target_size=8000):
    """
    balanced job title category + skill coverage for integrated sampling
    """

    print("=== start integrated strategic sampling ===")

    # 1. categorized job titles
    df_skill_cleaned['job_category'] = df_skill_cleaned['job_title'].apply(categorize_job)

    # 2. identifying critical skill
    print("analyzing critical skills...")
    skill_freq = {}
    for skills in df_skill_cleaned['skills_list']:
        for skill in skills:
            skill_freq[skill] = skill_freq.get(skill, 0) + 1

    total_jobs = len(df_skill_cleaned)

    # select top 70% skills - except unique skill
    important_skills = {skill for skill, freq in skill_freq.items()
                       if freq / total_jobs >= 0.005}

    print(f"num of importarnt skills: {len(important_skills)}")
    print(f"num of total skills: {len(skill_freq)}")

    # 3. calculate skill coverage
    def calculate_skill_coverage(job_skills):
        if not important_skills:
            return 0
        covered_skills = set(job_skills).intersection(important_skills)
        return len(covered_skills) / len(important_skills)

    df_skill_cleaned['coverage_score'] = df_skill_cleaned['skills_list'].apply(calculate_skill_coverage)

    # 4. sampling by category
    sampled_dfs = []
    category_counts = df_skill_cleaned['job_category'].value_counts()

    for category, total_count in category_counts.items():
        category_df = df_skill_cleaned[df_skill_cleaned['job_category'] == category]

        # calculate sample size by categorym proportion
        category_proportion = total_count / len(df_skill_cleaned)
        category_target_size = int(target_size * category_proportion)


        category_target_size = max(category_target_size, 50)

        print(f"category '{category}': num of {category_target_size} of total num {len(category_df)}")

        if len(category_df) <= category_target_size:

            sampled_dfs.append(category_df)
        else:
            # high skill coverage priority
            high_coverage = category_df[category_df['coverage_score'] >= 0.2]
            low_coverage = category_df[category_df['coverage_score'] < 0.2]

            # high coverage sample 70%
            high_sample_size = int(category_target_size * 0.7)
            if len(high_coverage) >= high_sample_size:
                high_sample = high_coverage.nlargest(high_sample_size, 'coverage_score')
            else:
                high_sample = high_coverage

            # low covarage sample 30%
            low_sample_size = category_target_size - len(high_sample)
            if len(low_coverage) >= low_sample_size:
                low_sample = low_coverage.sample(n=low_sample_size, random_state=42)
            else:
                low_sample = low_coverage

            category_sample = pd.concat([high_sample, low_sample], ignore_index=True)
            sampled_dfs.append(category_sample)

    # 5. integrated all samples
    integrated_sample = pd.concat(sampled_dfs, ignore_index=True)

    # 6. final check
    final_skills = set()
    for skills in integrated_sample['skills_list']:
        final_skills.update(skills)

    coverage = len(important_skills.intersection(final_skills)) / len(important_skills)

    print(f"\n=== integrated sampling result ===")
    print(f"final sample size: {len(integrated_sample)}")
    print(f"skill coverage: {coverage:.3f}")
    print(f"num of unique job title: {integrated_sample['job_title'].nunique()}")
    print(f"num of unique skill: {len(final_skills)}")
    print("\n distribution by category:")
    print(integrated_sample['job_category'].value_counts())

    return integrated_sample


df_integrated_sample = integrated_strategic_sampling(df_skill_cleaned, 8000)
```

```python
def efficient_data_merging(df_skill_sample, df_course_cleaned, max_merged_size=15000):
    """
    effiencint data merging with sample job data and course data

    """

    print("=== effienct data merging ===")

    # 1. make skill list from couse data
    df_course_cleaned['course_skills_list'] = df_course_cleaned['Skills'].str.split(', ')

    # 2. merging skill records
    merged_records = []

    # 코스 데이터를 스킬별로 인덱싱 (메모리 효율성)
    print("indexing course data...")
    skill_course_map = {}

    for _, course_row in df_course_cleaned.iterrows():
        if isinstance(course_row['course_skills_list'], list):
            for skill in course_row['course_skills_list']:
                if skill not in skill_course_map:
                    skill_course_map[skill] = []
                skill_course_map[skill].append({
                    'Title': course_row['Title'],
                    'Organization': course_row['Organization'],
                    'Ratings': course_row['Ratings'],
                    'course_url': course_row['course_url'],
                    'Duration': course_row['Duration'],
                    'Difficulty': course_row['Difficulty']
                })

    print(f"num of indexing skill: {len(skill_course_map)}")

    # 3. using chunk for save the memory
    chunk_size = 1000
    total_processed = 0

    for i in range(0, len(df_skill_sample), chunk_size):
        chunk = df_skill_sample.iloc[i:i+chunk_size]

        for _, job_row in chunk.iterrows():
            job_skills = job_row['skills_list']

            for skill in job_skills:
                if skill in skill_course_map:
                    for course_info in skill_course_map[skill]:
                        merged_records.append({
                            'job_link': job_row['job_link'],
                            'job_title': job_row['job_title'],
                            'company': job_row['company'],
                            'job_level': job_row['job_level'],
                            'job_type': job_row['job_type'],
                            'job_skills': job_row['job_skills'],
                            'skill': skill,
                            'course_title': course_info['Title'],
                            'course_organization': course_info['Organization'],
                            'course_rating': course_info['Ratings'],
                            'course_url': course_info['course_url'],
                            'course_duration': course_info['Duration'],
                            'course_difficulty': course_info['Difficulty']
                        })


            if len(merged_records) >= max_merged_size:
                print(f"merging limit: {max_merged_size}")
                break

        if len(merged_records) >= max_merged_size:
            break

        total_processed += len(chunk)
        print(f"process progress: {total_processed}/{len(df_skill_sample)}")

    # 4. creating merging dataframe
    merged_df = pd.DataFrame(merged_records)

    print(f"\n=== complete merging ===")
    print(f"size of final integrated data: {len(merged_df)}")
    print(f"unique job title: {merged_df['job_title'].nunique()}")
    print(f"unique course : {merged_df['course_title'].nunique()}")
    print(f"unique skill: {merged_df['skill'].nunique()}")

    return merged_df

    # save merged_df file to csv file
    merged_df.to_csv('merged_df.csv', index=False)
```
